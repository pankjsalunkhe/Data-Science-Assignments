Calculating VIF

```python
import pandas as pd
import statsmodels.formula.api as smf

variables = ['Age', 'KM', 'HP', 'CC', 'Drs', 'Grs', 'QT', 'WT']
vif_values = []

for var in variables:
    formula = f"{var} ~ {'+'.join(variables)} - {var}"
    rsquared = smf.ols(formula, data=df1).fit().rsquared
    vif = 1 / (1 - rsquared)
    vif_values.append(vif)

d = {'Variables': variables, 'VIF': vif_values}
d1 = pd.DataFrame(d)
print(d1)
```
